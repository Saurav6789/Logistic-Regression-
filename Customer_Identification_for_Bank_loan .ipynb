```python
#Cleaning the negative values in the Experience data set 

#Separating the negative values , positive values and people with zero experince respectively

posExp_df=bank_customer_df.loc[bank_customer_df["Experience"]>0]
negExp=bank_customer_df.Experience < 0
column_name = 'Experience'
neg_list=bank_customer_df.loc[negExp]['ID'].tolist() # getting the customer ID who has negative experience

```

```python
negExp.value_counts()
```

**Code to replace the negatiive values with the median of experience in the dataset**

It follows the following steps 

1. For the record with the ID , get the Age column
2. For the record with the ID , get the Education column
3. Filter the records matching the above criteria from the data frame which has records with positive experience and take the median
4. Apply the median back to the location which had negative experience



```python
for id in neg_list:
    age = bank_customer_df.loc[np.where(bank_customer_df['ID']==id)]["Age"].tolist()[0]
    education=bank_customer_df.loc[np.where(bank_customer_df['ID']==id)]["Education"].tolist()[0]
    filtered_df=posExp_df[(posExp_df.Age == age) & (posExp_df.Education == education)]
    exp=filtered_df["Experience"].median()
    bank_customer_df.loc[bank_customer_df.loc[np.where(bank_customer_df['ID']==id)].index, 'Experience'] = exp
```

Applying the models 

```python
#Splitting the dataset into training and test data set

X=pd.DataFrame(columns=['Age','Income','Family','CCAvg','Education','Mortgage','SecuritiesAccount','CDAccount','CreditCard','Online'],data=bank_customer_df)
y=bank_customer_df.iloc[:,9]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=100)


```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
logit =LogisticRegression()
logit.fit(X_train,y_train)
y_pred=logit.predict(X_test)
```

**Confusion Matrix for Logistic Regression model**

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
